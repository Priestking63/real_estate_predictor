```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_moscow = pd.read_csv('avito_apartmentsmoskva.csv')
df_piter = pd.read_csv('avito_apartmentssankt-peterburg.csv')
df_kazan = pd.read_csv('avito_apartmentskazan.csv')
df_nizniy = pd.read_csv('avito_apartmentsnizhniy_novgorod.csv')
df_novosib = pd.read_csv('avito_apartmentsnovosibirsk.csv')
df_ekb = pd.read_csv('avito_apartmentsekaterinburg.csv')
```

```python
df['Общая площадь'] = df['Общая площадь'].apply(lambda x: float(x.split()[0]))
df['Площадь кухни'] = df['Площадь кухни'].apply(lambda x: float(str(x).split()[0]))
```

```python
df = df.dropna(axis=1, thresh= len(df)/1.5)
```

```python
mask = df['Площадь кухни'].notna() & df['Общая площадь'].notna()
percentage = (df.loc[mask, 'Площадь кухни'] / df.loc[mask, 'Общая площадь']).mean()

print(f"Средний процент: {percentage:.2%}")

mask_na = df['Площадь кухни'].isna()
df.loc[mask_na, 'Площадь кухни'] = (df.loc[mask_na, 'Общая площадь'] * percentage).round(1)
```

```python
s_mean = df[df['Количество комнат'] == 'студия']['Общая площадь'].mean()
room_area_ratio = (df['Общая площадь'] - df['Площадь кухни']) / s_mean

mask = df['Количество комнат'] == 'свободная планировка'

df.loc[mask, 'Количество комнат'] = np.floor(room_area_ratio[mask]).astype(int)
```

```python
df['Количество комнат'] = df['Количество комнат'].replace('студия', 0)
df['Количество комнат'] = df['Количество комнат'].apply(lambda x: str(x).split(' ')[0])
```

```python
df['Ремонт'] = df.groupby('Город')['Ремонт'].transform(
    lambda x: x.fillna(x.mode()[0])
)
```

```python
df['Этажей в доме'] = df['Этаж'].apply(lambda x: int(x.split()[-1]))
df['Этаж'] = df['Этаж'].apply(lambda x: int(x.split()[0]))
```

```python
def smart_floor_feature(floor, total_floors):
    
    if total_floors == 1:
        return 0.5
    relative = (floor) / (total_floors)
    
    if total_floors <= 5:  # хрущевки и малоэтажки
        if floor == 1:
            return relative * 0.7  # штраф за первый этаж
        elif floor == total_floors:
            return relative * 1.1  # небольшой бонус за последний
        else:
            return relative
            
    else:  # высотки
        if floor == 1:
            return relative * 0.6  # большой штраф за первый
        elif floor == total_floors:
            return relative * 1.4  # большой бонус за пентхаус
        elif floor >= total_floors - 2:
            return relative * 1.2  # бонус за высокие этажи
        else:
            return relative

df['умный_относ_этаж'] = df.apply(lambda x: smart_floor_feature(x['Этаж'], x['Этажей в доме']), axis=1)
```

```python
df['Год постройки'] = df.groupby(['Город','Тип дома'])['Год постройки'].transform(
    lambda x: x.fillna(x.mean())
).round(0)
```

```python
def fill_na_smart(df, col, floor_col='Этажей в доме', group_col='Тип дома'):
    """Заполняет пропуски: 0 если <=5 этажей, иначе модой по группе"""
    mask_na = df[col].isna()
    mask_low = df[floor_col] <= 5
    
    df.loc[mask_na & mask_low, col] = 0
    df.loc[mask_na & ~mask_low, col] = df.loc[mask_na & ~mask_low, group_col].map(
        df.groupby(group_col)[col].agg(lambda x: x.mode()[0] if not x.mode().empty else 0)
    )
    return df
```

```python
df = fill_na_smart(df, 'Пассажирский лифт')
df = fill_na_smart(df, 'Грузовой лифт')
df[['Пассажирский лифт','Грузовой лифт']] = df[['Пассажирский лифт', 'Грузовой лифт']].replace('нет', 0)
```

```python
df = df[df['Цена'] > df['Цена'].quantile(0.01)]
df = df[df['Цена'] < df['Цена'].quantile(0.95)]
rename_dict = {
    'Цена': 'price',
    'Ссылка': 'link',
    'Общая площадь': 'total_area',
    'Количество комнат': 'rooms',
    'Площадь кухни': 'kitchen_area',
    'Этаж': 'floor',
    'Ремонт': 'renovation',
    'Тип дома': 'house_type',
    'Пассажирский лифт': 'passenger_lift',
    'Грузовой лифт': 'cargo_lift',
    'Парковка': 'parking',
    'Город': 'city',
    'Этажей в доме': 'floors_total',
    'умный_относ_этаж': 'smart_floor_ratio',
    'возраст_дома': 'house_age',
    'период_строительства': 'construction_period',
    'Год постройки' : 'build_year'
}

df = df.rename(columns=rename_dict)
```

```python
def create_city_specific_features(df):
    """Создает признаки учитывающие специфику городов"""
    
    # Питерские исторические дома (особый случай)
    df['is_spb_historical'] = ((df['city'] == 'Питер') & 
                              (df['house_age'] > 100) & 
                              (df['house_type'])).astype(int)
    
    df = df.drop('build_year', axis = 1)
    
    return df
def create_city_aware_age_features(df):
    """Создает возрастные признаки с учетом специфики городов"""
    df = df.copy()
    
    # Базовый возраст
    df['house_age'] = 2026 - df['build_year']
    
    def get_city_specific_age_group(city, age):
        if city == 'Питер':
            if age <= 5: return 'спб_новостройка'
            elif age <= 20: return 'спб_современная'
            elif age <= 60: return 'спб_советская'
            elif age <= 100: return 'спб_дореволюционная'
            else: return 'спб_историческая'
        elif city == 'Москва':
            if age <= 5: return 'мск_новостройка'
            elif age <= 20: return 'мск_современная'
            elif age <= 60: return 'мск_советская'
            elif age <= 100: return 'мск_старая'
            else: return 'мск_историческая'
        else:  # Другие города
            if age <= 5: return 'др_новостройка'
            elif age <= 20: return 'др_современная'
            elif age <= 60: return 'др_советская'
            else: return 'др_старая'
    
    df['city_specific_age_group'] = df.apply(
        lambda x: get_city_specific_age_group(x['city'], x['house_age']), axis=1
    )
    
    df['age_city_premium'] = df.apply(calculate_age_city_premium, axis=1)
    return df

def calculate_age_city_premium(row):
    """Рассчитывает премию за возраст в зависимости от города"""
    city = row['city']
    age = row['house_age']
    
    if city == 'Питер':
        if age > 100: return 1.2  
        elif age > 80: return 1.1
        elif age <= 5: return 1.5
        else: return 1.0
    elif city == 'Москва':
        if age <= 5: return 1.6 
        elif age > 80: return 1.1
        else: return 1.0
    else:  # Другие города
        if age <= 5: return 1.3
        elif age > 50: return 0.9 
        else: return 1.0

df = create_city_aware_age_features(df)
df = create_city_specific_features(df)
```

```python
# Если хотите использовать энкодер или что-то кроме катбуст
import pickle


class SafeCategoricalEncoder:
    def __init__(self):
        self.encoding_maps = {}
        self.fitted_columns = []
    
    def fit(self, df, categorical_cols, target):
        self.encoding_maps = {}
        self.fitted_columns = categorical_cols
        
        for col in categorical_cols:
            n_categories = df[col].nunique()
            
            if n_categories <= 5:
                means = df.groupby(col)[target].mean().sort_values()
                self.encoding_maps[col] = {
                    'type': 'ordered_label',
                    'mapping': {cat: i for i, cat in enumerate(means.index)}
                }
            else:
                global_mean = df[target].mean()
                group_means = df.groupby(col)[target].mean()
                group_counts = df.groupby(col)[target].count()
                
                self.encoding_maps[col] = {
                    'type': 'bayesian',
                    'global_mean': global_mean,
                    'group_means': group_means.to_dict(),
                    'group_counts': group_counts.to_dict()
                }
        return self
    
    def transform(self, df):
        df_encoded = df.copy()
        
        for col, encoding_info in self.encoding_maps.items():
            if col in df.columns:
                if encoding_info['type'] == 'ordered_label':
                    df_encoded[col] = df[col].map(encoding_info['mapping'])
                else:
                    global_mean = encoding_info['global_mean']
                    group_means = encoding_info['group_means']
                    group_counts = encoding_info['group_counts']
                    
                    def encode_value(x):
                        if x in group_means:
                            count = group_counts.get(x, 0)
                            return (group_means[x] * count + global_mean * 50) / (count + 50)
                        return global_mean
                    
                    df_encoded[col] = df[col].apply(encode_value)
        
        return df_encoded
    
    def fit_transform(self, df, categorical_cols, target):
        return self.fit(df, categorical_cols, target).transform(df)
    
    def save(self, path):
        with open(path, 'wb') as f:
            pickle.dump(self, f)
    
    @classmethod
    def load(cls, path):
        with open(path, 'rb') as f:
            return pickle.load(f)
```

```python
categorical_cols = df.drop('link', axis = 1).select_dtypes(include=['object', 'category']).columns.tolist()
encoder = SafeCategoricalEncoder()
df_enc = encoder.fit_transform(df, categorical_cols, 'price')
encoder.save('encoder.pkl')
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# VIF для проверки мультиколлинеарности
def calculate_vif(df, numerical_cols):
    X = add_constant(df[numerical_cols].dropna())
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

vif_results = calculate_vif(df.drop('link', axis = 1), numerical_cols)
print("VIF факторы:")
print(vif_results.sort_values('VIF', ascending=False))
```

```python
def generate_eda_report(df_enc, target_col='price'):
    """Генерирует комплексный отчет по данным"""
    
    print("=" * 50)
    print("КОМПЛЕКСНЫЙ АНАЛИЗ ДАННЫХ")
    print("=" * 50)
    
    print(f"Размер датасета: {df_enc.shape}")
    print(f"Пропуски: {df_enc.isna().sum().sum()}")
    print(f"Дубликаты: {df_enc.duplicated().sum()}")
    
    print(f"\nЦЕЛЕВАЯ ПЕРЕМЕННАЯ ({target_col}):")
    print(f"Мин: {df_enc[target_col].min():.2f}")
    print(f"Макс: {df_enc[target_col].max():.2f}")
    print(f"Среднее: {df_enc[target_col].mean():.2f}")
    print(f"Медиана: {df_enc[target_col].median():.2f}")
    
    numerical_cols = df_enc.select_dtypes(include=['int64', 'float64']).columns
    if target_col in numerical_cols:
        correlations = df_enc[numerical_cols].corr()[target_col].sort_values(ascending=False)
        print(f"\nТОП-10 корреляций с {target_col}:")
        print(correlations.head(10))
    
    categorical_cols = df_enc.select_dtypes(include=['object', 'category']).columns
    print(f"\nКАТЕГОРИАЛЬНЫЕ ПЕРЕМЕННЫЕ ({len(categorical_cols)}):")
    for col in categorical_cols:
        print(f"{col}: {df_enc[col].nunique()} уникальных значений")

generate_eda_report(df)
```
